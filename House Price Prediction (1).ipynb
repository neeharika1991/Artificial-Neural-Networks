```python
# Feature
X = data1.drop('price',axis=1)

# Labeling
y = data1['price']

# Spliting the data into train and test set
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
scaler= MinMaxScaler()

# fiting and transfrom
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# everything has been scaled between 1 and 0
print('Maximum: ',X_train.max())
print('Minimum: ',X_train.min())
```

```python
model = Sequential()

# input layer for with the help of Dense
model.add(Dense(19,activation='relu'))

# seconf and hidden layers with relu activation function
model.add(Dense(19,activation='relu'))

model.add(Dense(19,activation='relu'))

model.add(Dense(19,activation='relu'))

# output layer
model.add(Dense(1))

# model compilation
model.compile(optimizer='adam',loss='mse')
```

```python
# fitting the data into model
model.fit(x=X_train,y=y_train.values,validation_data=(X_test,y_test.values),batch_size=128,epochs=400)
```

```python
loss = pd.DataFrame(model.history.history)  # graph of loss

plt.figure(figsize=(15,5))
sns.lineplot(data=loss,lw=3)
plt.xlabel('Epoch')
plt.ylabel('')
plt.title('Training_Loss per Epoch')
sns.despine()
```

```python
# predictions on the test set
predictions = model.predict(X_test)

print('MAE: ',mean_absolute_error(y_test,predictions))
print('MSE: ',mean_squared_error(y_test,predictions))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,predictions)))
print('Variance Regression Score: ',explained_variance_score(y_test,predictions))

print('\n\nDescriptive Statistics:\n',data1['price'].describe())
```

```python
f, axess = plt.subplots(1, 2,figsize=(15,5))

# Prediction with model
plt.scatter(y_test, predictions)

# Perfect predictions
plt.plot(y_test,y_test,'r')

errors = y_test.values.reshape(6484, 1) -predictions
sns.distplot(errors, ax=axess[0])

sns.despine(left=True, bottom=True)
axess[0].set(xlabel='Error', ylabel='', title='Error_Histogram')
axess[1].set(xlabel='Test True Y', ylabel='Model_Predictions', title='Model Prediction vs Perfect Fiting')
```
